```python
from matplotlib import pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
from scipy.spatial import ConvexHull
```

```python
# Define the 8 custom vertices manually
points = np.load("verts.npy")
# Compute the convex hull
hull = ConvexHull(points)

# Create 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the vertices
ax.scatter(points[:, 0], points[:, 1], points[:, 2], color='blue', s=50)

# Plot the convex hull faces
for simplex in hull.simplices:
    face = points[simplex]
    poly = Poly3DCollection([face], alpha=0.5, facecolor='lightblue', edgecolor='blue')
    ax.add_collection3d(poly)

# Set plot limits
ax.set_xlim([-150, 150])
ax.set_ylim([-150, 150])
ax.set_zlim([-100, 150])

# Set labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.title('3D Convex Polytope Defined by Linear Constraints')
plt.tight_layout()
plt.savefig('../data/plots/convex_polytope_example.png')
```
