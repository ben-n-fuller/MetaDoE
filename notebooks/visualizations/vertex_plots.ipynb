```python
# Load vertices
verts = np.load("../data/coetzer_verts_3.npy")
# plot_polyhedron(verts, 'Reparameterized Polyhedron', '../data/reparameterized_polyhedron.png')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import ConvexHull
from matplotlib.patches import Polygon
```

```python
verts[:, 0]
```

```python
hull = ConvexHull(verts)
hull_points = verts[hull.vertices]  # Ordered points along convex hull

# Plotting
fig, ax = plt.subplots()

# Add the convex hull polygon
poly = Polygon(hull_points, closed=True, facecolor='lightblue', edgecolor='black', alpha=0.6)
ax.add_patch(poly)

# Remove ticks
ax.set_xticks([])
ax.set_yticks([])

# Optionally scatter the original vertices
ax.scatter(verts[:, 0], verts[:, 1], color='red', zorder=5)
ax.set_aspect('equal')

plt.title('Reparameterized Mixture Constraints')

plt.show()

plt.savefig('coetzer_reparameterized_constraints.png')
```
