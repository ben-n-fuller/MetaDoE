```python
from lib.plot import plot_scores
import numpy as np
import itertools
```

```python
objectives = ['ackley', 'rastrigin', 'griewank', 'rosenbrock']
enforcers = ['Penalty', 'Parametric']
combos = list(itertools.product(objectives, enforcers))

for (objective, enforcer) in combos:
    # Load the data
    data = np.load(f'../data/10kd_{objective}_{enforcer}.npy')
    scores = data['scores']
    
    # Plot the data
    plot_scores(scores, 1, 200, f"{objective.capitalize()} Fitness - {enforcer.capitalize()} Enforcement - 3D", f"../data/plots/10kd_{objective}_{enforcer}")
```
